```python
import pandas as pd
df=pd.read_csv('weather-cities.csv')
df
```

```python
df.pivot(index='day',columns='city')
```

```python
df.pivot(index='day',columns='city', values='temperature')
```

```python
df.pivot(index='temperature',columns='city')
```

```python
df=pd.DataFrame({
    'date':['5/1/2017','5/1/2017','5/2/2017','5/2/2017','5/1/2017','5/1/2017','5/2/2017','5/2/2017'],
    'city':['New York','New York','New York','New York','Dhaka','Dhaka','Dhaka','Dhaka'],
    'temperature':[65,51,70,72,75,78,82,80],
    'humidity':[56,54,60,62,80,83,85,26]
})
df
```

```python
df.pivot_table(index='city',columns='date',aggfunc='count')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='sum')
```

```python
df.pivot_table(index='city',columns='date',margins=True)
```
