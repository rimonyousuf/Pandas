```python
import pandas as pd

bd_weather=pd.DataFrame({
    'city':['Dhaka','Khulna','Meherpur'],
    'temperature':[32,45,30],
    'humidity':[80,60,78]
})
bd_weather
```

```python
import pandas as pd

us_weather=pd.DataFrame({
    'city':['New York','Chicago','Orlando'],
    'temperature':[32,14,35],
    'humidity':[68,65,75]
})
us_weather
```

```python
df=pd.concat([bd_weather,us_weather])
df
```

```python
df=pd.concat([bd_weather,us_weather],ignore_index=True)
df
```

```python
df=pd.concat([bd_weather,us_weather], keys=['Bangladesh','United States'])
df
```

```python
df.loc['Bangladesh']
```

```python
df.loc['United States']
```
